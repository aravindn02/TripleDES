```python
def hex2bin(pt):
    mp = {'0': "0000",
          '1': "0001",
          '2': "0010",
          '3': "0011",
          '4': "0100",
          '5': "0101",
          '6': "0110",
          '7': "0111",
          '8': "1000",
          '9': "1001",
          'A': "1010",
          'B': "1011",
          'C': "1100",
          'D': "1101",
          'E': "1110",
          'F': "1111",
          'a': "1010",
          'b': "1011",
          'c': "1100",
          'd': "1101",
          'e': "1110",
          'f': "1111"}
    bin=""
    for i in range(len(pt)):
        bin=bin+mp[pt[i]]


    return bin
def bin2dec(binary):
 
    binary1 = binary
    decimal, i, n = 0, 0, 0
    while(binary != 0):
        dec = binary % 10
        decimal = decimal + dec * pow(2, i)
        binary = binary//10
        i += 1
    return decimal
 
# Decimal to binary conversion
 
def dec2bin(num):
    res = bin(num).replace("0b", "")
    if(len(res) % 4 != 0):
        div = len(res) / 4
        div = int(div)
        counter = (4 * (div + 1)) - len(res)
        for i in range(0, counter):
            res = '0' + res
    return res

def bin2hex(s):
    mp = {"0000": '0',
          "0001": '1',
          "0010": '2',
          "0011": '3',
          "0100": '4',
          "0101": '5',
          "0110": '6',
          "0111": '7',
          "1000": '8',
          "1001": '9',
          "1010": 'A',
          "1011": 'B',
          "1100": 'C',
          "1101": 'D',
          "1110": 'E',
          "1111": 'F'}
    hex = ""
    for i in range(0, len(s), 4):
        ch = ""
        ch = ch + s[i]
        ch = ch + s[i + 1]
        ch = ch + s[i + 2]
        ch = ch + s[i + 3]
        hex = hex + mp[ch]
 
    return hex
```

```python
def hexconverter(ascii):
 
    
    hexout = ""
 

    for i in range(len(ascii)):
 
        ch = ascii[i]

 
       
        in1 = ord(ch)
    
     
        part = hex(in1).lstrip("0x").rstrip("L")
 
     
        hexout += part
 

    return hexout
```

```python
def comparator(c,d):
    x=[]
    for i in range(28):
        x.append(0)
    bits=0
    while bits<=27:
        if c[bits]==d[bits]:
            x[bits]=0
        else:
            x[bits]=1 
        bits=bits+1
    cmp=""
    for i in x:
        cmp=cmp+str(i)
    return cmp
```

```python
def xor(a, b):
    ans = ""
    for i in range(len(a)):
        if a[i] == b[i]:
            ans = ans + "0"
        else:
            ans = ans + "1"
    return ans
```

```python
def ls(x,n):
    #print("Entered left shift function")
    l=""
    for i in range(n,len(x)):
        l=l+x[i]
    for i in range(n):
        l=l+x[i]
    #print("Left shift value:"+l)
    return l
```

```python
def rs(x,n):
    r=""
    for i in range(0,len(x)-n):
        r=r+x[i]
    for i in range(len(x)-n,len(x)):
        r=x[i]+r
    #print(f"After right shift of {n} bits:{r}")
    return r
```

```python
def permute(k, arr, n):

    permutation = ""

    for i in range(0, n):

        permutation = permutation + k[arr[i] - 1]

    return permutation
```

```python
def keygen(k,new_key):
    k=hex2bin(k)
    k=permute(k,pc1,56)
    left=k[0:28]
    right=k[28:56]
    keylist=[] #binary
    keyhex=[] #hex
    vls=""
    vrs=""
    for m in range(0,16):
        left = ls(left, shift_table[m])
        right = ls(right, shift_table[m])
        cmp=comparator(left,right)
        if left[0]=='0':
            vrs=rs(left,1)
        elif left[0]=="1":
            vrs=rs(left,2)
        
        if cmp[0]=="0":
            vls=ls(cmp,1)
        elif cmp[0]=="1":
            vls=ls(cmp,2) 
        
        left=vls #vrs
        right=vrs #vls
        combine_key=vrs+vls #vrs+vls
        
        rkey=permute(combine_key,pc2,48)
        rkey=xor(rkey,new_key)
        keylist.append(rkey)
        keyhex.append(bin2hex(rkey))
    return keyhex,keylist
```

```python
def encryption(pt, kbin, khex):
    pt = hex2bin(pt)
    
    # IP Table
    pt = permute(pt, iptable, 64)
    left = pt[0:32]
    right = pt[32:64]
    for i in range(0, 16):
        #  convert 32 bits to 48 bits
        rexp = permute(right, exptable, 48)
        
        #rexp=xor(rexp,new_key)

        xor_x = xor(rexp, kbin[i])
 
        # S-box
        sbox_out = ""
        for j in range(0, 8):
            r = bin2dec(int(xor_x[j * 6] + xor_x[j * 6 + 5]))
            c = bin2dec(
                int(xor_x[j * 6 + 1] + xor_x[j * 6 + 2] + xor_x[j * 6 + 3] + xor_x[j * 6 + 4]))
            val = sbox[j][r][c]
            sbox_out = sbox_out + dec2bin(val)
 
        # Straight D-box
        sbox_out = permute(sbox_out, permtable, 32)
        # XOR left and sbox_str
        result = xor(left,sbox_out)
        left=result
        
        if(i != 15):
            left, right = right, left
    combine = left + right
    #fp table
    cipher_text=permute(combine,fptable,64)
    return cipher_text
```
